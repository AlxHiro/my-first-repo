```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Microsoft JhengHei'  # 或 'SimHei' / 'Arial Unicode MS'
plt.rcParams['axes.unicode_minus'] = False  # 讓負號正常顯示

# 讀取 CSV
df = pd.read_csv("../data/titanic.csv")
```

```python
# 性別人數
sns.countplot(x='Sex', data=df)
plt.title("性別分布")
plt.show()

# 生還人數
sns.countplot(x='Survived', data=df)
plt.title("生還 vs 死亡 數量")
plt.show()
```

```python
df.groupby('Sex')['Survived'].mean()
df.groupby('Pclass')['Survived'].mean()
df.groupby(['Sex', 'Pclass'])['Survived'].mean()
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("不同性別的生還率")
plt.show()
```

```python
sns.histplot(data=df, x='Age', hue='Survived', bins=30, kde=True)
plt.title("年齡分布與生還情形")
plt.show()

sns.boxplot(data=df, x='Survived', y='Fare')
plt.title("票價 vs 生還與否")
plt.show()
```

```python
sns.violinplot(data=df, x='Pclass', y='Age', hue='Survived', split=True)
plt.title("艙等 vs 年齡 vs 生還情形")
plt.show()
```
